```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ARRAY
# a=cv2.imread("top_left.jpg")
# b=cv2.imread("top_right.jpg")
# c=cv2.imread("bottom_left.jpg")
# d=cv2.imread("bottom_right.jpg")
# e=cv2.imread("center.jpeg")

a="top_left.jpg"
b="top_right.jpg"
c="bottom_left.jpg"
d="bottom_right.jpg"
e="center.jpeg"

m=11
array=np.array([a,b,c,d,e], dtype='object')

for i in array:
    print(i)
    x=cv2.imread(i)
    x=cv2.cvtColor(x,cv2.COLOR_BGR2RGB)
    x=cv2.resize(x,(200,200))
    
    if (m==11):
        m=np.all(x)
        count=1
    elif (count==1):
        n=np.all(x)
        count+=1
    elif (count==2):        
        o=np.all(x)
        count+=1
    elif (count==3):
        p=np.all(x)
        count+=1
    else:
        q=np.all(x)
        
    plt.axis("off")
    plt.imshow(x)
    plt.show()
    print()
```

```python
#Border for the Base Images (Square)

border_h=np.zeros((10,430,3),dtype="object")
border_v=np.zeros((200,10,3),dtype="object")

h1=np.hstack([border_v,m,border_v,n,border_v])
h2=np.hstack([border_v,o,border_v,p,border_v])
v=np.vstack([border_h,h1,border_h,h2,border_h])

#Showing Collage of base 4 Images with border

plt.imshow(v)
plt.axis('off')
plt.title(v.shape)
plt.show()

#Border for Center Image and Combining it with image

border_h=np.zeros((10,220,3),dtype='object')
border_v=np.zeros((200,10,3),dtype='object')
h2=np.hstack([border_v,q,border_v])
center=np.vstack([border_h,h2,border_h])
plt.imshow(center)
plt.show()

#Replacing the centre part of collage with center image

v[105:325,105:325,:]=center

plt.imshow(v)
plt.show()
```

```python
# Reducing the size and converting the Color Scheme

top_left=cv2.imread("top_left.jpg")
top_left=cv2.cvtColor(top_left,cv2.COLOR_BGR2RGB)
top_left=cv2.resize(top_left,(200,200))
plt.axis("off")
plt.imshow(top_left)
plt.show()

top_right=cv2.imread("top_right.jpg")
top_right=cv2.cvtColor(top_right,cv2.COLOR_BGR2RGB)
top_right=cv2.resize(top_right,(200,200))
plt.axis("off")
plt.imshow(top_right)
plt.show()

center=cv2.imread("center.jpeg")
center=cv2.cvtColor(center,cv2.COLOR_BGR2RGB)
center=cv2.resize(center,(200,200))
plt.axis("off")
plt.imshow(center)
plt.show()

bottom_left=cv2.imread("bottom_left.jpg")
bottom_left=cv2.cvtColor(bottom_left,cv2.COLOR_BGR2RGB)
bottom_left=cv2.resize(bottom_left,(200,200))
plt.axis("off")
plt.imshow(bottom_left)
plt.show()

bottom_right=cv2.imread("bottom_right.jpg")
bottom_right=cv2.cvtColor(bottom_right,cv2.COLOR_BGR2RGB)
bottom_right=cv2.resize(bottom_right,(200,200))
plt.axis("off")
plt.imshow(bottom_right)
plt.show()
```
